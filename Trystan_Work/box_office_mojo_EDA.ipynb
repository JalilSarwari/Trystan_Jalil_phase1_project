```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df =(pd.read_csv('zippedData/bom.movie_gross.csv.gz', delimiter=','))
df.head(15)
```

```python
df.isna()
df.dropna(subset=['foreign_gross'], inplace = True)
df['foreign_gross'] = df['foreign_gross'].apply(lambda x: x.replace(",", ""))
df['foreign_gross'] = df['foreign_gross'].apply(lambda x: float(x))
```

```python
df.reset_index(inplace = True)
```

```python
df.sort_values(by=['domestic_gross'], ascending = False, inplace = True)
df.reset_index(inplace = True)
```
