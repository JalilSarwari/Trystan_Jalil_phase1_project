Dataset:
movie_budgets

```python
df4 =(pd.read_csv('/Users/weatherford/Documents/Flatiron/Trystan_Jalil_phase1_project/zippedData/tn.movie_budgets.csv.gz', delimiter=','))
df4.tail(100)
```

```python
df4['worldwide_gross'] = df4['worldwide_gross'].apply(lambda x: x.replace(",", ""))
df4['worldwide_gross'] = df4['worldwide_gross'].apply(lambda x: x.replace("$", ""))
df4['worldwide_gross'] = df4['worldwide_gross'].apply(lambda x: int(x))
```

```python
df4['domestic_gross'] = df4['domestic_gross'].apply(lambda x: x.replace(",", ""))
df4['domestic_gross'] = df4['domestic_gross'].apply(lambda x: x.replace("$", ""))
df4['domestic_gross'] = df4['domestic_gross'].apply(lambda x: int(x))
```

```python
df4['production_budget'] = df4['production_budget'].apply(lambda x: x.replace(",", ""))
df4['production_budget'] = df4['production_budget'].apply(lambda x: x.replace("$", ""))
df4['production_budget'] = df4['production_budget'].apply(lambda x: int(x))
```

```python
df4["total_gross"] = df4["production_budget"] + df4["domestic_gross"]
```

```python
df4.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(10,4))

#by default, line plot. First argument is X list/array.
# Second argument is Y list/array.
ax.scatter(df4["production_budget"], df4["total_gross"], alpha = 0.3) 

ax.set_xlabel('Production Budget')
ax.set_ylabel('Total Gross')
ax.set_title('Production Budget & Total Gross')

plt.tight_layout()
plt.show()
```
